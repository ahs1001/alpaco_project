## <작업>

```python
import pandas as pd

df_kyobo = pd.read_excel("교보문고_종합_베스트셀러_상품리스트-2012~2022.xlsx")

print(type(df_kyobo), end = '\n\n') # 데이터 타입
print(df_kyobo.shape, end = '\n\n') # 데이터 크기
print(df_kyobo.info(), end = '\n\n') # 요약정보
df_kyobo.head(3)
```

```python
# 순위열 중복 확인
print(True in (df_kyobo['순위'].value_counts() > 1), end = '\n\n') # --> False: 중복 없음


# "판매상품ID"를 기준으로 중복 된 값 제거
df_kyobo_1 = df_kyobo.drop_duplicates(subset='판매상품ID', keep='first')
print(df_kyobo_1.shape, end = '\n\n')


# "판매상품ID"를 기준으로 오름차순 정렬
df_kyobo_1  = df_kyobo_1.sort_values(by = '판매상품ID')

# []'정가', '판매가' .'적릴율', '할인율' , '적립예정포인트'] 열의 천단위 구분기호 (,), '%' 제거 후 int로 타입 변겅
df_kyobo_1['정가'] = df_kyobo_1['정가'].str.replace(',', '').astype(int)
df_kyobo_1['판매가'] = df_kyobo_1['판매가'].str.replace(',', '').astype(int)
df_kyobo_1['할인율'] = df_kyobo_1['할인율'].str.replace('%', '').astype(int)
df_kyobo_1['적립율'] = df_kyobo_1['적립율'].str.replace('%', '').astype(int)
df_kyobo_1['적립예정포인트'] = df_kyobo_1['적립예정포인트'].str.replace(',', '').astype(int)

# '발행(출시)일자' 열 (format : 2023Y12m9d)
df_kyobo_1['발행(출시)일자'] = pd.to_datetime(df_kyobo_1['발행(출시)일자'], format = '%Y%m%d')

# '인물' 열 이름을 '저자'오 변경
df_kyobo_1 = df_kyobo_1.rename(columns = {'인물' : '저자'})

# 열 순서 변경
# 변경 열 순서 : ['순위', '판매상품ID', '상품코드', '상품명', '저자', 출판사', '분야','발행(출시)일자', '정가', '판매가', '할인율', '적립율', '적립예정포인트']
df_kyobo_1 = df_kyobo_1[['순위', '판매상품ID', '상품코드', '상품명', '저자', '출판사', '분야','발행(출시)일자', '정가', '판매가', '할인율', '적립율', '적립예정포인트']]

# 덮어 쓰기 및 인덱스 초기화
df_kyobo_1 =  df_kyobo_1.reset_index(drop = True) # drop=True # 인덱스 새로 추가
df_kyobo = df_kyobo_1

# 도서 분약 분류 확인
print(df_kyobo['분야'].value_counts(), '\n\n')

print(df_kyobo.info(), end = '\n\n')
df_kyobo.head(3)
```

```python
# .pkl로 저장
import pickle

df_kyobo.to_pickle('df_kyob.pkl')
```
